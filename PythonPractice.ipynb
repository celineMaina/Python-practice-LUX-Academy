```python
# Write a program that counts how many times each word appears in a list of words: ["apple", "banana", "apple", "cherry", "banana", "banana"].
fruits = ["apple", "banana", "apple", "cherry", "banana", "banana"]
# Create a dictionary
fruits_count = {}

for fruit in fruits:
  if fruit in fruits_count:
    fruits_count[fruit] += 1
  else:
    fruits_count[fruit] = 1

print(fruits_count)
```

```python
# Ask user to input three numbers and store them in a list. Then, convert the list to a set and check if any duplicates were removed; tell the user if there were duplicates.
numbers = []

#Loop 3 times to enter 3 numbers
for x in range(3):
  num = int(input("Enter a number: "))
  numbers.append(num)

numbers_set = set(numbers)

if len(numbers_set) < len(numbers):
  print('Duplicates were removed.')
else:
  print('There were no duplicates')

print('Original list: ', numbers)
print('New set: ', numbers_set)
```

```python
# Given a list of numbers, ask the user to input the numbers, then compute: number of positive, negative, and zero values using conditionals.
numbers = []

for n in range(5):
    num = int(input(f"Enter number {n+1}: "))
    numbers.append(num)

print(numbers)

for x in numbers:
    if x > 0:
      print(x, 'is Positive')
    elif x == 0:
      print(x, 'is Zero')
    else:
      print(x, 'is Negative')

positive = 0
negative = 0
zero = 0

for y in numbers:
    if y > 0:
      positive += 1
    elif y == 0:
      zero += 1
    else:
      negative += 1

print('There are', positive, 'positive numbers in this list')
print('There are', negative, 'negative numbers in this list')
print('There are', zero, 'zeroes in this list')

```

```python
# Write a small program that asks the user to input their name, age, and a list of 3 favorite fruits (comma-separated).
# Store their data in a dictionary like {"name": ..., "age": ..., "fruits": [...]} and then print a summary: “Name: …, Age: …, Fruits: …”.
name = input('What is your name?: ')
age = input('How old are you?: ')
fruits = input('Enter your 3 favorite fruits, separated by commas: ')

user_info = {}
user_info["Name"] = name
user_info["Age"] = age
user_info["Fruits"] = [f.strip() for f in fruits.split(",")]

print(f"Name: {user_info['Name']}, Age: {user_info['Age']}, Fruits: {user_info['Fruits']}")
```
