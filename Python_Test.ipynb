# Loops(for/while)


```python
# Use a for loop to print numbers from 1 to 10.
for num in range(1, 11):
  print(num)
```

```python
# Use a while loop to print numbers from 5 down to 1.
num = 5
while num >=1:
  print(num)
  num -= 1
```

```python
# Loop through a list of names and print each one.
names = ["Peter", "Sandra", "Meg", "Chris"]

for name in names:
  print(name)
```

```python
# Loop through a dictionary and print each key and value.
user = {"name":"Bob", "age":31}

for x, y in user.items():
  print(x, ':', y)
```

```python
# Create a list of 5 numbers and use a for loop to compute the sum.
numbers = [10, 20, 30, 40, 50]
total = 0

for num in numbers:
  total += num
print(f'Sum of numbers is {total}')
```

# Conditional Statements

```python
# Ask the user for their age and print "Adult" if age ≥ 18.
age = int(input("Enter your age: "))

if age >= 18:
  print("Adult")
```

```python
# Write a program that checks if a number is even or odd.
number = int(input("Enter a number: "))

if number%2 == 0:
  print("Even")
else:
  print("Odd")
```

```python
# If temp > 30, print "Hot", else print "Cool".
temp = int(input("Enter the temparature: "))

if temp > 30:
  print("Hot")
else:
  print("Cool")
```

```python
# Write an if/elif/else to check grade range:
# ≥ 80 → "A"
# ≥ 60 → "B"
# else → "C"
grade = int(input("Enter student grade: "))

if grade >= 80:
  print("A")
elif 60 <= grade < 80:
  print("B")
else:
  print("C")
```

```python
# Ask the user for a password and check if it matches "12345".
password =str(input("enter your password: "))

if password == "12345":
  print("Logged in successfully!")
else:
  print("Sorry, wrong password.")
```

## Sets

```python
# Create a set of 5 numbers.
numbers = {10, 20, 30, 40, 50}
print(numbers)
```

```python
# Add an element to the set.
numbers.add(True)
print(numbers)
```

```python
# Remove an element from the set.
numbers.remove(20)
print(numbers)
```

```python
# Show the difference between two sets A and B.
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

diff = a.difference(b)
# diff = a - b #works the same way
print(diff)
```

## Dictionaries

```python
# Create a dictionary called student with keys: name, age, grade.
student = {"name": 'Linda', "age": 23, "grade": 'A'}
print(student)
```

```python
# Access the value of the "name" key in the dictionary.
print(student["name"])
```

```python
# Change the "age" value to a new number.
student.update({"age": 21})
print(student)
```

```python
# Add a new key "gender" to the dictionary.
student["gender"] = 'Female'
print(student)
```

```python
# Remove the "grade" key from the dictionary.
student.pop("grade")
print(student)
```

```python
# Loop through the dictionary keys and print each one.
for x in student:
  print(x)
```

```python
# Loop through the dictionary values and print each one.
for y in student.values():
  print(y)
```

```python
# Print all key-value pairs using .items().
for z in student.items():
  print(z)
```

```python
# Use in to check if "age" exists in the dictionary.
if "age" in student:
  print('"age" is a key in the dictionary')
```

```python
# Create a nested dictionary representing two students
nested_students = {
    "student1": {"name": 'Kai', "age": 20, "gender": 'Male'},
    "student2": {"name": 'Sally', "age": 19, "gender": 'Female'}
}

print(nested_students)
```
